### Load and Save Data (Streamlined!!)

```python
def load_and_save(coin, meme=False):
    if not meme:
        command = f'npx binance-fetch -d 2021-01-01 2024-10-31 -p spot -t klines -s {coin} -i 1h -o rawdata'
    else:
        # Only gather testing data
        command = f'npx binance-fetch -d 2024-01-01 2024-01-02 -p spot -t klines -s {coin} -i 1h -o rawdata'
    subprocess.run(command, shell=True)

    subprocess.run('unzip -o "rawdata/*.zip" -d rawdata/', shell=True)
    os.system('rm rawdata/*.zip')

    data = pd.DataFrame()
    for file in os.listdir('rawdata'):
        if file.endswith('.csv'):
            df = pd.read_csv(f'rawdata/{file}', header=None)
            data = pd.concat([data, df], axis=0)

    cols = ['Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close time', 'Quote asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore']
    data.columns = cols
    data = data.sort_values('Open time')
    data = data.reset_index(drop=True)

    if ("Ignore" in data.columns):
        if (data['Ignore'].sum() == 0):
            data = data.drop(columns=['Ignore'])
    
    data['LogReturns'] = np.log(data['Close']/data['Close'].shift(1))
    data.loc[0, 'LogReturns'] = 0

    if not meme:
        data.to_csv(f'data/{coin}.csv', index=False)
    else:
        data.to_csv(f'memedata/{coin}.csv', index=False)
    os.system('rm rawdata/*.csv')
```

```python
for coin in ['shibusdt', 'solusdt', 'ethusdt', 'btcusdt', 'adausdt', 'dotusdt', 'maticusdt', 'xrpusdt', 'dogeusdt', 'lunausdt']:
    if not os.path.exists(f'data/{coin}.csv'):
        load_and_save(coin)
```

```python
for coin in ['pepeusdt', 'bonkusdt', 'wifusdt', 'flokiusdt']:
    if not os.path.exists(f'memedata/{coin}.csv'):
        load_and_save(coin, meme=True)
```
